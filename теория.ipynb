#ROC AUC
Пример как считать
Алгоритм бинарной классификации выдаёт значения bi
, принадлежащие отрезку [0,1]. Всего имеется 10000 наблюдений. Если ранжировать их по возрастанию bi
, то окажется, что наблюдения с yi=1
 занимают ровно места с 6501 по 6600. Найдите площадь под ROC-кривой. Ответ округлите до сотых.

 Решение:
 b(x) - прогнозы, y - ответы
 Первые 6500 - это 0
 Следующие 100 - это 1
 Оставшиеся 3400 - снова 0
 Для построения рок кривой делаем по убыванию
 3400 нулей, 100 единиц, 6500 нулей
 Горизонталь делим на столько частей, сколько у нас ноликов (3400 + 6500=9900 кусочков), вертикаль 100 кусочков.
 Далее по шагам сверху вниз строим кривую. Получается
 3400 шагов по нулям вправо, 100 вверх(на 1), 6500 вправо

 Считаем площадь = высота * ширину = 1 * (6500/9900) ~~ 0,66



 В общем случае:

 1) Сортируем ответы от большего к меньшему

 2) Нам не очень важны значения, важны 1 и 0(метки)!

 3) строим квадрат 1 на 1

 4) 0 по горизонтали, 1 по вертикали

 5) Идём по делениям также сверху вниз

 6) Считаем площадь под рок кривой. Высота * ширину

#Функция потерь - ошибка модели

ROC AUC
Доля верных предсказаний - это точность  = TP / (TP + FP)
TPR = recall  = TP / (TP + FN) - доля верных предсказаний относительно всего положительного класса.
То есть было 5 пациентов ABCDE, из них A и D больны, модель предсказала ABC, тогда TPR = 1 / 2. Аналогично с FPR = FP / (FP + TN)

#ошибка модели. Bias-variance decomposition
смещение +  разброс + шум
смещение - насколько хорошо модель предсказывает целевую переменную, чем больше смещение, тем хуже модель предсказывает целевую пер.
разброс - показывает степень переобученности модели, чем больше разброс, тем больше модель переобучена
err = bias^2 + var(дисперсия) + ню^2 (только по данным определяется, обычно на него не смотрим)

Например: линейная регрессия - слабая модель, потому что смещение очень большое

У случайного леса ТАКОЕ ЖЕ СМЕЩЕНИЕ как у дерева
Классификация объектов проводится путём голосования деревьев внутри случайного леса
Случайный лес не переобучается с ростом числа деревьев
Случайный лес обучается с помощью bootstrap, но в этих выборках могут быть повторяющиеся объекты
В случайном лесе при выборе наилучшего разбиения в вершине перебирается лишьс лучайное подмножество признаков
